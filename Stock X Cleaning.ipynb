```python
#Library Imports
import numpy as np
import pandas as pd
```

```python
#Import CSV file and check first 5 records
original = pd.read_csv(r'C:\Users\mathe\Desktop\SneakerProject\StockX_Sneakers.csv')
df = original
df.head()
```

```python
#Checking Null Values in dataset
df.isnull().sum()
#Checking columns for any type of null values
df.isnull().any()
```

```python
#Checking Data Types of our Dataset
df.dtypes
```

```python
#Changing the dtype of 'Release Date' column into a datetime
df['Release Date'] =  pd.to_datetime(df['Release Date'], format='%m/%d/%Y')
df.head()
```

```python
#Changing the dtype of 'Order Date' column into a datetime
df['Order Date'] =  pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df.head()
```

```python
#Replacing the '-' in the "Sneaker Name" column to a Space
df['Sneaker Name'] = df['Sneaker Name'].str.replace('-',' ')
df.head()
```

```python
#Getting rid of the ',' and '$' in the Sale Price
df['Sale Price'] = df['Sale Price'].str.replace('$','')
df['Sale Price'] = df['Sale Price'].str.replace(',','')
```

```python
#Same as above but in the Retail Price column.
df['Retail Price'] = df['Retail Price'].str.replace('$','')
df['Retail Price'] = df['Retail Price'].str.replace(',','')
```

```python
#Check the Data
df.head(10)
```

```python
#Export the dataset out
df.to_csv(r'C:\Users\mathe\Desktop\SneakerProject\CleanedSneaker.csv')
```
